### Accelerometer Data Graphing

This notebook demonstrates the Vibes module, which makes it convenient to load CSV and TAB delimited 
time series data and allows graphing of time and frequency (FFT) graphs.

```python
# Setup programming environment
import vibes
from pylab import *
%matplotlib inline
```

```python
# Load data from text file

filename = 'shelf_normal.txt'
dataset1 = vibes.load_ni_txt_file(filename)


#Show 1st column of dataset
dataset1[0].time_series.times
#dataset1[0].time_series.values
```

```python


# Set name of each column and plot graph
dataset1[0].name = "On table, std"
dataset1[0].color = "red"   # blue, green, red, cyan, magenta, yellow, black, white

dataset1[1].name = "Above load cell, std"
dataset1[1].color = "grey"

plt1 = vibes.plot_time_series(dataset1)

plt1 = vibes.plot_spectrum(dataset1, xlim=[0,200], ylim=[1e-10,1e-3])

plt1 = vibes.plot_binned_spectrum(dataset1, xlim=[0,200], ylim=[1e-10,1e-3])
show()  #Show plots
```

```python
# Create seperate plots of second experiment

filename = 'shelf_inverted.txt'
dataset2 = vibes.load_ni_txt_file(filename)

# Set name of each column and plot graph
dataset2[0].name = "On table, inverted"
dataset2[0].color = "red"   # blue, green, red, cyan, magenta, yellow, black, white

dataset2[1].name = "Above load cell, inverted"
dataset2[1].color = "green"

plt1 = vibes.plot_time_series(dataset2[0])
plt1 = vibes.plot_time_series(dataset2[1])
show()  #Show plots
```

```python
dataset2[1].color = "red"

# Build list with 2nd column of each dataset
data2plot = [ dataset1[1], dataset2[1] ]  

plt1 = vibes.plot_time_series(data2plot)
plt1 = vibes.plot_spectrum(data2plot, xlim=[0,200], ylim=[1e-10,1e-3])
plt1 = vibes.plot_binned_spectrum(data2plot, xlim=[0,200], ylim=[1e-10,1e-3])
show()  #Show plots
```
