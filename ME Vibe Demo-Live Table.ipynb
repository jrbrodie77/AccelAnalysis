### Accelerometer Data Graphing

This notebook demonstrates the Vibes module, which makes it convenient to load CSV and TAB delimited 
time series data and allows graphing of time and frequency (FFT) graphs.

```python
# Setup programming environment

import vibes
from pylab import *
%matplotlib inline
```

```python
# Load data from text file

filename = 'desk_trial.txt'
dataset1 = vibes.load_ni_txt_file(filename)
```

```python


# Set name of each column and plot graph
dataset1[0].name = "On table, ref"
dataset1[0].color = "red"   # blue, green, red, cyan, magenta, yellow, black, white

dataset1[1].name = "On table, main position"
dataset1[1].color = "grey"

plt1 = vibes.plot_time_series(dataset1)
plt1 = vibes.plot_spectrum(dataset1, xlim=[0,200], ylim=[1e-7,1e-2])
plt1 = vibes.plot_binned_spectrum(dataset1, xlim=[0,200], ylim=[1e-7,1e-2])
show()  #Show plots
```

```python
tf = dataset1[1].calc_tf(dataset1[0])
```

```python
vibes.plot_tf( dataset1[1], xlim=[0,200], ylim=[-30, 30] )
show()
```
