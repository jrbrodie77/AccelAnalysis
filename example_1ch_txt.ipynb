### Accelerometer Data Graphing

This notebook demonstrates the Vibes module. It opens a single column of magnitude only time data.  The time series is generated using the dt: value in the header of the file (Brett's exported format)

```python
# Setup programming environment

import vibes
from pylab import *
%matplotlib inline
```

```python
# Load data from text file

filename = "warehouse_table1_up-down_nofan_010715.txt"
dataset1 = vibes.load_1col_txt_file(filename)
```

```python
# Set name of each column and plot graph
dataset1[0].name = "On table, ref"
dataset1[0].color = "red"   # blue, green, red, cyan, magenta, yellow, black, white

plt1 = vibes.plot_time_series(dataset1)
plt1 = vibes.plot_spectrum(dataset1, xlim=[0,200], ylim=[1e-10,1e-6])
plt1 = vibes.plot_binned_spectrum(dataset1, xlim=[0,200], ylim=[1e-10,1e-6])
show()  #Show plots
```
